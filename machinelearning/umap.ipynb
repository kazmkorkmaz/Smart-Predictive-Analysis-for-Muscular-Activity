```python
import numpy as np
import umap
from sklearn.preprocessing import StandardScaler
import time
import matplotlib.pyplot as plt
import matplotlib.colors as clr
from collections import Counter
```

```python
x = np.loadtxt("DNN_features_June20.txt")
y = np.loadtxt("DNN_WeightLevel_Labels_June20.txt")
print(x.shape)
```

```python
reducer = umap.UMAP()
```

```python
scaled_data = StandardScaler().fit_transform(x)
print(scaled_data.shape)

starting = time.time()
embedding = reducer.fit_transform(scaled_data)
timePassed = time.time() - starting
print("Time passed for training: ", timePassed, " seconds")
print(embedding.shape)

red = "red"
blue = "blue"
yellow = "yellow"
colorList = []
#print(y)
for t in y:
    if(t == 0):
        colorList.append(red)
    elif(t==1):
        colorList.append(blue)
    elif(t == 2):
        colorList.append(yellow)
        
plt.scatter(embedding[:,0],embedding[:,1],c=colorList,marker="o", picker=True)
plt.title("UMAP for Scaled Data (Weight Level Labels)")
```

```python
# SMOTE oversampling codes
from imblearn.over_sampling import SMOTE

x = np.loadtxt("DNN_features_June20.txt")
y = np.loadtxt("DNN_WeightLevel_Labels_June20.txt")
print(x.shape)
counter = Counter(y)
print(counter)

oversample = SMOTE()
x,y = oversample.fit_resample(x,y)
print("After oversampling...")
print("X = ", x.shape)
print("y = ", y.shape)
counter = Counter(y)
print(counter)

scaled_data = StandardScaler().fit_transform(x)

starting = time.time()
embedding = reducer.fit_transform(x)
timePassed = time.time() - starting
print("Time passed for training: ", timePassed, " seconds")
print(embedding.shape)
red = "red"
blue = "blue"
yellow = "yellow"
colorList = []
#print(y)
for t in y:
    if(t == 0):
        colorList.append(red)
    elif(t==1):
        colorList.append(blue)
    elif(t == 2):
        colorList.append(yellow)
plt.scatter(embedding[:,0],embedding[:,1],c=colorList,marker="o", picker=True)
plt.title("UMAP After Oversampling (Weight Level Labels)")
```
